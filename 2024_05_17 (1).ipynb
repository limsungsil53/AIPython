```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rc('font', family='NanumGothic')

df2 = pd.DataFrame({'도시': ['서울', '부산', '인천', '대구'],
                   '인구 수': [9904312, 3448737, 2890451, 2466052]})

plt.bar(df2['도시'], df2['인구 수'])
plt.title('도시별 인구 수')
plt.xlabel('도시')
plt.ylabel('인구 수')
plt.show()
```

```python
import pandas as pd

penguins_data=pd.read_csv('penguins.csv')
penguins_data.info()
sns.barplot(x='species', y='body_mass_g', data=penguins_data, hue='sex')

sns.pairplot(penguins_data,hue='species')
plt.show()

sns.regplot(x='body_mass_g', y='flipper_length_mm', data=penguins_data)
```
